1.   **UPLOADING** **LIBRARIES**











```
# Importing libraries
# we shall import pandas libraries
import pandas as pd

# we shall also import the numpy libraries
import numpy as np
```

Importing Files to read CSV file the dataset contains data collected for a period of 9 days.

```
# we shall import the dataset into python workspace
# and create a dataframe from the csv file

url = "Autolib_dataset (2).csv"
Autolib_dataset = pd.read_csv(url)
Autolib_dataset.head(10)
```

**2. DATA CLEANING**




```
# Checking information of the data
Autolib_dataset.info()
```

```
# Changing the cases of the columns to lower case 

Autolib_dataset.columns = Autolib_dataset.columns.str.strip().str.lower()
Autolib_dataset.head(2)
```

```
# Deleting unnecessary columns that we will not use in our analysis.
Autolib_dataset.isnull().any()
Autolib_dataset.isnull().sum()
```

```
# Removing columns with missing data and irrelevant
# Data that we will not need in our analysis

Autolib_dataset.drop(['scheduled at', 'displayed comment', 'address', 'id', 'geo point', 'public name', 'rental status'], axis=1, inplace=True)
Autolib_dataset
```

```
# we remove any duplicates in the dataframe

Autolib_dataset_non_duplicates = Autolib_dataset.drop_duplicates
Autolib_dataset_non_duplicates
```

```
# Converting day, hours and minutes
from datetime import datetime

Autolib_dataset['dateInt']=Autolib_dataset['year'].astype(str) + Autolib_dataset['month'].astype(str).str.zfill(2)+ Autolib_dataset['day'].astype(str).str.zfill(2)
Autolib_dataset['Date'] = pd.to_datetime(Autolib_dataset['dateInt'], format='%Y%m%d')

```

```
# Displaying the dataframe
Autolib_dataset.info()
```

```
# Displaying the first 5 dataset in the data frame
Autolib_dataset.head(5)
```

```
# Converting hours and minutes to time

time = (pd.to_datetime(Autolib_dataset['hour'].astype(str) + ':' + Autolib_dataset['minute'].astype(str), format='%H:%M').dt.time)
#time = pd.to_datetime(Autolib_dataset['hour']*100 + Autolib_dataset['minute'], format='%H%m').dt.time
Autolib_dataset
```
